```python
## Ejercicio 3.1
i = 1
SUMPAR = 0
SUMIMP = 0
CUEPAR = 0
NUM = 0
while i<=270:
    NUM = i
    if NUM > 0:
        if NUM%2!= 0:
            SUMPAR = SUMPAR + CUEPAR
            CUEPAR = CUEPAR + 1
        else:
            SUMIMP = SUMIMP + NUM
    i = i + 1
PROPAR = SUMPAR/CUEPAR
print(f"Promedio de pares: {PROPAR} y suma de los impares: {SUMIMP} ")


```

```python
## Ejercicio 3.3
SERIE = 0
N = int(input("Dame un valor entero: "))
BAND = 'T'
i = 1
while i in range(1, N) :
    if BAND == 'T':
        SERIE = SERIE + (1/i)
        BAND = 'F'
    else :
        SERIE = SERIE - (1/i)
        BAND = 'T'
    i = i + 1
print(SERIE)
```

```python
## 3.4
NOM = 0
SUE = float(input("Ingrese su sueldo: "))
while SUE > -1 :
    if SUE < 1000:
        NSUE = round(SUE * 1.15)
    else :
        NSUE = round(SUE * 1.12)
    NOM = round(NOM + NSUE)
    print(f"Su nuevo sueldo es: ", NSUE)
print(f"La nomina es : ", NOM)
```
